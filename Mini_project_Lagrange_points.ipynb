**Question 1:-**

**Solution 1:-**

Using Newton law of gravitation and second law of motion,  we can write the net force on the satellite of mass 'm', located at position (x,y) as-

$$ m\vec{a} = \vec{F_{S}} + \vec{F_{E}} $$   

where, 

$$ \vec{Fs} = \text{gravitational force act on satellite by sun} $$

$$ \vec{Fe} = \text{gravitational force act on satellite by earth} $$



$$ $$
Miniproject Fig 1.JPG

$$ $$
In the above figure we have taken the origin of coordinate system is at center of mass of earth-sun system. In this way, location of earth is ($ R_{E} $ ,0 ) ,location of sun is ($ -R_{S} $ , 0 ) and location of our satellite is (x , y ).

$$ $$

Now after resolving the component of net force on satellite along x and y direction, we get,

Net Force along x-direction-
$$ ma_{x} = -F_{S}\cos(\alpha) -F_{E}\cos(\beta) $$

Net Force along y-direction-
$$ ma_{y} = -F_{S}\sin(\alpha) -F_{E}\sin(\beta) $$

where, 
$$ F_{S} = \frac{GM_{S}m}{ (y^2 + ( x+R_{S} )^2 )^2} $$

$$ F_{E} = \frac{GM_{E}m}{( y^2 + ( x-R_{E} )^2 )^2} $$

$$ \tan(\alpha) = \frac{y}{( x + R_{S} )}$$

$$ \tan(\beta) = \frac{y}{( x - R_{E} )}$$

$$ $$

After putting these into equation (2) and (3), we finally get this,

$$ $$

$$ a_{x} =-\frac{GM_{S}( x + R_{S} )}{( y^2 + ( x+R_{S} )^2 )^{3/2}} -\frac{GM_{E}( x - R_{E} )}{( y^2 + ( x-R_{E} )^2 )^{3/2}} $$


$$ a_{y} =-\frac{GM_{S}y}{( y^2 + ( x+R_{S} )^2 )^{3/2}} -\frac{GM_{E}y}{( y^2 + ( x-R_{E} )^2 )^{3/2}} $$

$$ $$

As we know, Lagrange points are locations where our satellite can rotate synchronously with earth, as the earth rotates around Sun. Therefore,
$$  $$ 
$$ a_{x} = -\omega^{2} x $$
$$ a_{y} = -\omega^{2} y $$

where,
$$ \omega = \text{Angular velocity of earth rotating around sun in circular orbit} $$

$$ $$

Now to find the lagrange points we have to find the solutions of below equations-

$$ $$

$$ -\frac{GM_{S}( x + R_{S} )}{( y^2 + ( x+R_{S} )^2 )^{3/2}} -\frac{GM_{E}( x - R_{E} )}{( y^2 + ( x-R_{E} )^2 )^{3/2}} + \omega^{2} x = 0 $$

$$ $$

$$ -\frac{GM_{S}y}{( y^2 + ( x+R_{S} )^2 )^{3/2}} -\frac{GM_{E}y}{( y^2 + ( x-R_{E} )^2 )^{3/2}} + \omega^{2} y = 0 $$

Let,
$$ f_{1}(x,y) = -\frac{GM_{S}( x + R_{S} )}{( y^2 + ( x+R_{S} )^2 )^{3/2}} -\frac{GM_{E}( x - R_{E} )}{( y^2 + ( x-R_{E} )^2 )^{3/2}} + \omega^{2} x $$

$$ $$

$$ f_{2}(x,y) = -\frac{GM_{S}y}{( y^2 + ( x+R_{S} )^2 )^{3/2}} -\frac{GM_{E}y}{( y^2 + ( x-R_{E} )^2 )^{3/2}} + \omega^{2} y $$

$$ $$

Thus we have to solve,

$$ f_{1}(x,y) = 0 $$
$$ f_{2}(x,y) = 0 $$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from scipy import optimize
from scipy.optimize import fsolve

def func1( x , y ):

  rat = 333000    # rat = mass of sun divided by mass of earth

  G = 1.0         # We refactorised the various constants according to the scale of the problem

  m_sat = 1.0

  F_x = m_sat*(-(G*rat * (x + (1/(rat+1))) / ( y**2 + ( x + (1/(1+rat)) )**2 )**(3/2) )
   - (G* (x - rat/(rat + 1)) / ( y**2 + ( x - (rat/(rat+1)) )**2 )**(3/2) ) + G*x*(rat + 1))

  return F_x 

def func2( x , y ):

  rat = 333000

  G = 1.0

  m_sat = 1.0

  F_y = m_sat*(-(G*rat * y / ( y**2 + ( x + (1/(rat+1)) )**2 )**(3/2) ) - (G* y / ( y**2 +
        ( x - (rat/(rat+1)) )**2 )**(3/2) ) +  G*y*(rat + 1))

  return  F_y



xval = np.linspace(0.980, 1.020, 100)
yval = np.linspace(-0.020, 0.020, 100)

#------Creating 2-D grid of features-------------
[X, Y] = np.meshgrid( xval , yval )
  
 
#--Z1 = Effective force in x direction , Z2 = Effective force in y direction
Z1 = func1( X , Y )  
Z2 = func2( X , Y )  

#--Z = Net Effective force 
Z =np.sqrt( Z1**2 + Z2**2 )

#------Fx plot--------------

plt.figure(figsize=(10,10))

c1 = plt.pcolormesh( X , Y , Z1, cmap = 'seismic', vmin = Z1.min()/2e2 , vmax = Z1.max()/2e2 )
c1bar = plt.colorbar(c1)
plt.title('Effective force in x direction')
plt.xlabel('x Distance( AU )')
plt.ylabel('y Distance( AU )')

plt.show()

#------Fy plot--------------

plt.figure(figsize=(10,10))

c2 = plt.pcolormesh( X , Y , Z2 , cmap = 'seismic', vmin = Z2.min()/2e2 , vmax = Z2.max()/2e2)
c1bar = plt.colorbar(c2)
plt.title('Effective force in y direction')
plt.xlabel('x Distance( AU )')
plt.ylabel('y Distance( AU )')

plt.show()

#------Net force plot--------------

plt.figure(figsize=(10,10))
c = plt.pcolormesh( X , Y , Z , cmap = 'hot', vmin = 0.0 , vmax = Z.max()/4.8e2 )
c1bar = plt.colorbar(c)
plt.title('Net Effective force')
plt.xlabel('x Distance( AU )')
plt.ylabel('y Distance( AU )')

plt.show()

```

```python
#------FINDING L1 , L2 , L3 (using single variable Newton Method----------------------------------------------------

import time

start_newton = time.time()

x_init = [ -5.010 , 0.985 , 1.000 ]   # X_init = [ L1 , L2 , L3 ]

x_next = x_init

epilson = 1e-8

def dF_x( x ):

   rat = 333000

   G = 1.0

   m_sat = 1.0

   derivative = m_sat*( (G*rat * 2 / (  x + (1/(1+rat)) )**3 ) + (G*2 / ( ( x - (rat/(rat+1)) )**3 )) + G*(rat + 1))

   return derivative


def func3( x ):
  
   rat = 333000

   G = 1.0

   m_sat = 1.0

   fx = m_sat*(-(G*rat / ( x + (1/(1+rat)) )**2  ) - (G /  ( x - (rat/(rat+1)) )**2  ) + G*x*(rat + 1))   

   return fx

for i in range( 0 , 3):
   
   F_x = func3( x_next[i] )

   iter = 0

   while np.abs( F_x ) > epilson :

      x_next[i] = x_next[i] - F_x / dF_x( x_next[i] )

      F_x = func3( x_next[i] )

      iter = iter + 1

   print(f"Number of iterations done for L{i+1} :",iter)
   print(f"Location of L{i+1} : x =  " , x_next[i] )
   print(f"Net force expected at L{i+1} : F_expected = " , F_x )
   print("Actual force at this point : F_actual =" , func3( x_next[i] ))
   print(" ")

end_newton = time.time()

print("The time of execution of Newton Raphson is :", end_newton-start_newton)
```

```python
#------------------ Finding L4 and L5 using multivariable Newton Mrthod -----------------------#

import numpy as np

rat = 333000

G = 1.0

m_sat = 1.0

def jacobian(xy):      # Creating the Jacobian matrix of the system of equations

    x, y = xy

    return [[-1.0*(2.99999099101804 - 3.0*x)*(1.0*x - 0.999996997006015)/(y**2 + (x - 0.999996997006015)**2)**2.5 - 1.0*(-3.0*x - 9.00898195500914e-6)*(333000.0*x + 0.999996997006015)/(y**2 + (x + 3.00299398500305e-6)**2)**2.5 - 1.0/(y**2 + (x - 0.999996997006015)**2)**1.5 - 333000.0/(y**2 + (x + 3.00299398500305e-6)**2)**1.5 + 333001.0
             , 3.0*y*(1.0*x - 0.999996997006015)/(y**2 + (x - 0.999996997006015)**2)**2.5 + 3.0*y*(333000.0*x + 0.999996997006015)/(y**2 + (x + 3.00299398500305e-6)**2)**2.5],
            [-1.0*y*(2.99999099101804 - 3.0*x)/(y**2 + (x - 0.999996997006015)**2)**2.5 - 333000.0*y*(-3.0*x - 9.00898195500914e-6)/(y**2 + (x + 3.00299398500305e-6)**2)**2.5
             , 3.0*y**2/(y**2 + (x - 0.999996997006015)**2)**2.5 + 999000.0*y**2/(y**2 + (x + 3.00299398500305e-6)**2)**2.5 - 1.0/(y**2 + (x - 0.999996997006015)**2)**1.5 - 333000.0/(y**2 + (x + 3.00299398500305e-6)**2)**1.5 + 333001.0]]


def function(xy):       # Given simultaneous equations in reduced form

    x, y = xy
    f1_x = m_sat*(-(G*rat * (x + (1/(rat+1))) / ( y**2 + ( x + (1/(1+rat)) )**2 )**(3/2) ) - (G* (x - rat/(rat + 1)) / ( y**2 + ( x - (rat/(rat+1)) )**2 )**(3/2) ) + G*x*(rat + 1))
    f2_x = m_sat*(-(G*rat * y / ( y**2 + ( x + (1/(rat+1)) )**2 )**(3/2) ) - (G* y / ( y**2 + ( x - (rat/(rat+1)) )**2 )**(3/2) ) +  G*y*(rat + 1))
    return [f1_x, f2_x]

# Newton method for finding the roots

def iterative_newton(fun, x_init, jacobian):

    max_iter = 500
    epsilon = 1e-8

    x_last = x_init

    for k in range(max_iter):
        # Solve J(xn)*( xn+1 - xn ) = -F(xn):
        J = np.array(jacobian(x_last))
        F = np.array(fun(x_last))

        difference = np.linalg.solve( J, -F )

        x_last = x_last + difference

        # Stop condition:
        if np.linalg.norm(difference) < epsilon:
            print('Newton Raphson convergence!, Number of iterations:', k )
            break

    else: # only if the for loop end 'naturally'
        print('not converged')

    return x_last


# For our 2 equations:
x_sol_L4 = iterative_newton(function, [5, 5], jacobian)    # Initial condition for L4
print(f"\nThe [x,y] corrdintaes of L4 are [{x_sol_L4[0]},{x_sol_L4[1]}] \n")

x_sol_L5 = iterative_newton(function, [-2, -2], jacobian)    # Initial condition for L5
print(f"\nThe [x,y] corrdintaes of L5 are [{x_sol_L5[0]},{x_sol_L5[1]}] \n")
```

```python
## Finding Lagrange Points using Optimize Libraray ##

from scipy.optimize import fsolve

start_fsolve = time.time()

def f2(guess):
    rat = 333000
    G = 1.0
    m_sat = 1.0
    x = guess[0]
    y = guess[1]
    p = np.zeros(2)
    p[0] = m_sat*(-(G*rat * (x + (1/(rat+1))) / ( y**2 + ( x + (1/(1+rat)) )**2 )**(3/2) ) 
        - (G* (x - rat/(rat + 1)) / ( y**2 + ( x - (rat/(rat+1)) )**2 )**(3/2) ) + G*x*(rat + 1))
    
    p[1] = m_sat*(-(G*rat * y / ( y**2 + ( x + (1/(rat+1)) )**2 )**(3/2) ) - (G* y / 
        ( y**2 + ( x - (rat/(rat+1)) )**2 )**(3/2) ) +  G*y*(rat + 1))
    
    return p

print("Lagrange Points derived using Scipy.Optimize\n")
L1 = fsolve(f2, [-2, 0])
print(f"The location of L1 in [x, y] coordinates is [ {L1[0]}, {L1[1]} ]\n")

L2 = fsolve(f2, [1.0, 0])
print(f"The location of L2 in [x, y] coordinates is [ {L2[0]}, {L2[1]} ]\n")

L3 = fsolve(f2, [0.8, 0])
print(f"The location of L3 in [x, y] coordinates is [ {L3[0]}, {L3[1]} ]\n")

L4 = fsolve(f2, [0.6, 1])
print(f"The location of L4 in [x, y] coordinates is [ {L4[0]}, {L4[1]} ]\n")

L5 = fsolve(f2, [0.6, -1])
print(f"The location of L5 in [x, y] coordinates is [ {L5[0]}, {L5[1]} ]\n")

end_fsolve = time.time()
print("\nThe time of execution using Fsolve is :", end_fsolve-start_fsolve)
print("\n\n\n\n\n")




```
